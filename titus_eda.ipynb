Here it is important to note that scipy does a two tailed T-Test by default and if one wants to obtain the simulated results of a one tailed T-Test then you have to divide the p-value in half.

Here we have split our data into matches in which there were no red cards, and games in which there were red cards.  From these matches we have taken random samples; this is because the group in which there were red cards was considerably smaller than the group in which there were no red cards, and the variance was likely to be so different as to be incomparable between the two groups. The poplulation for these groups is all premier league games for the 2018/2019 however with more sampling it would be possible to imagine the population being the differences between red card games and non red card games for all premier league seasons, or indeed all football matches. 

For the analysis of statistical significance in this case we have used the Welch's T-Test. We have done so for the following reasons:

    - These are two disparate groups; although the same teams might be playing in more than one match in which there is a red card, we are not making a paired comparison between repeated performance of specific teams. 
    - The variance between the two groups is not the same. The variance for the red card games is 2.115 and whereas the variance for the sample with no reds is 2.106. A Welch's T-Test is typically preferred when variances between samples are not equal
    - The Welch's T-Test is more robust. It is more capable of dealing with skewed distributions, and the number of goals in a game is a positively skewed distribution. 

We retrieved the critical t-value from a t-table, given an alpha value of 0.05. Our **t-value of 1.398** was lower than the **critical value of 1.672**. Given this information we conclude that we cannot reject our null hypothesis. 

```python
# Calculate degrees of freedom for welch test
def welch_dof(a, b):
    s1 = a.var(ddof=1)
    s2 = b.var(ddof=1)
    n1 = len(a)
    n2 = len(b)
    numerator = (s1/n1 + s2/n2)**2
    denominator = (s1/ n1)**2/(n1 - 1) + (s2/ n2)**2/(n2 - 1)
    return numerator/denominator
df = welch_dof(red_games_sample.goals, no_red_games_sample.goals)
df
```
